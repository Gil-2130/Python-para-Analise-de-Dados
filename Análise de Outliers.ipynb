## Análise de Valores Extremos usando métodos Univariados

```python
# Carregando bibliotecas
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from matplotlib import rcParams
from sklearn import datasets

%matplotlib inline
rcParams['figure.figsize'] = 8, 7

iris = datasets.load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
df.boxplot()
```

```python
iris_out = larg_cepa < 2.05
df[iris_out]
```

## aplicando a identificação de Outlier de Tukey

```python
pd.options.display.float_format = '{:.2f}'.format
X_df = pd.DataFrame(df['sepal width (cm)'])
sumario = X_df.describe()
print(sumario)
```

#### Calculando os limites inferior e superior do intervalo entre quartis

```python
iqr = sumario.loc['75%'] - sumario.loc['25%']
iqr
```

```python
limit_inf = sumario.loc['25%'] -iqr * 1.5
limit_inf
```

```python
limit_sup = sumario.loc['75%'] + iqr * 1.5
limit_sup
```

## Usando DBSCAN

```python
# importando bibliotecas
from matplotlib import rcParams
from sklearn.cluster import DBSCAN
from collections import Counter

%matplotlib inline
rcParams['figure.figsize'] = 8, 7
sns.set_style('whitegrid')

df = sns.load_dataset("iris")
cols = ['comp. da cepa', 'larg. da cepa', 'comp. da pétala', 'larg. da pétala', 'espécie']
df.columns = cols
X = df[cols[:4]].values
y = df['espécie'].values
```

#### Treinando o modelo para identificar outliers

```python
modelo = DBSCAN(eps=0.8, min_samples=19).fit(X)
print(modelo)
```

#### Visualizando resultados

```python
outliers_df = pd.DataFrame(X)
print(Counter(modelo.labels_))
```

```python
filtro = modelo.labels_ == -1
print(outliers_df[filtro])
```

```python
cores = modelo.labels_

plt.scatter(X[:,2], X[:,1], c=cores, s=120)
plt.xlabel('Comprimento da Pétala')
plt.ylabel('Largura da Cepa')
plt.title('DBSCAN para detecção de Outliers')
plt.show()
```
